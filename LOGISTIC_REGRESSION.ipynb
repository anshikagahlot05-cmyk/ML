Question 1: What is Simple Linear Regression (SLR)? Explain its purpose.

Answer:

Simple Linear Regression (SLR) is a statistical and supervised machine learning technique used to study the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line to the data.

Detailed Explanation:

Simple Linear Regression assumes that the relationship between X and Y is linear.

It uses historical data to learn the relationship between the variables.

The model represents this relationship using a straight-line equation.

The independent variable is also called the predictor or input variable.

The dependent variable is also called the response or output variable.

SLR helps in understanding how changes in X affect Y.

It is widely used in statistics, economics, business, and machine learning.

The regression line is also known as the best-fit line.

The goal is to minimize prediction error.

It helps in both analysis and prediction.

Purpose of Simple Linear Regression:

To find whether a relationship exists between two variables.

To identify the direction of the relationship (positive or negative).

To estimate the strength of the relationship.

To predict future values of the dependent variable.

To explain real-world phenomena using mathematical models.

To support decision-making based on data.

To analyze trends and patterns.

To reduce uncertainty in predictions.

To summarize large datasets.

To serve as a foundation for more complex regression models.

Question 2: What are the key assumptions of Simple Linear Regression?

Answer:

Simple Linear Regression is based on several assumptions that must be satisfied for the model to be valid and reliable.

Key Assumptions:

Linearity:
The relationship between independent and dependent variables must be linear.

Independence of Observations:
Each observation should be independent of others.

Homoscedasticity:
The variance of errors should remain constant for all values of X.

Normality of Errors:
Residuals should be normally distributed.

No Multicollinearity:
Only one independent variable should be present.

No Outliers:
Extreme values should not disproportionately influence the model.

Correct Model Specification:
The model should include all relevant variables.

Zero Mean of Errors:
The average of error terms should be zero.

Measurement Accuracy:
Variables should be measured accurately.

Random Sampling:
Data should be randomly sampled from the population.

Question 5: What is the method of least squares in linear regression?

Answer:

The method of least squares is a mathematical technique used to find the best-fit regression line by minimizing prediction errors.

Detailed Explanation:

The error is the difference between actual and predicted values.

Errors are also called residuals.

Each error is squared to remove negative signs.

All squared errors are summed.

The goal is to minimize this total sum.

This method finds optimal values of slope and intercept.

It ensures the smallest possible overall error.

It improves prediction accuracy.

It provides an objective optimization criterion.

It is the most commonly used estimation technique.

Question 6: What is Logistic Regression? How does it differ from Linear Regression?

Answer:

Logistic Regression is a classification algorithm used to predict binary outcomes.

Detailed Explanation:

It predicts probabilities between 0 and 1.

Uses a sigmoid (logistic) function.

Commonly used in classification problems.

Output is converted into class labels.

Used in spam detection, disease prediction, etc.

Differences Between Linear and Logistic Regression:

Linear Regression predicts continuous values.

Logistic Regression predicts categorical outcomes.

Linear Regression uses a straight line.

Logistic Regression uses an S-shaped curve.

Linear Regression uses MSE loss.

Logistic Regression uses log loss.

Linear output is unbounded.

Logistic output is bounded between 0 and 1.

Linear Regression solves prediction problems.

Logistic Regression solves classification problems.

Question 7: Name and briefly describe three common evaluation metrics for regression models.

Answer:

Mean Absolute Error (MAE):

Average of absolute prediction errors.

Easy to interpret.

Less sensitive to outliers.

Mean Squared Error (MSE):

Average of squared prediction errors.

Penalizes large errors.

Useful for optimization.

Root Mean Squared Error (RMSE):

Square root of MSE.

Same unit as target variable.

More interpretable.

Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

Answer:

from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2, 4, 6, 8, 10])

model = LinearRegression()
model.fit(X, Y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)


Output:

Slope: 2.0
Intercept: 0.0
