Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.

Answer:

Anomaly Detection is the process of identifying data points, patterns, or events that deviate significantly from normal behavior in a dataset.

Purpose of Anomaly Detection:

Identify unusual behavior.

Detect fraud or faults.

Improve system reliability.

Prevent losses.

Enable early warnings.

Types of Anomalies:
1. Point Anomalies

A single data point that is significantly different from others.

Example:

A credit card transaction of ₹5,00,000 when normal spending is ₹2,000.

Extremely high taxi fare compared to normal fares.

Key Points:

Most common anomaly type.

Easy to detect.

Treated independently.

Works well with statistical methods.

Used in fraud detection.

2. Contextual Anomalies

An anomaly that is abnormal in a specific context (time, location, season).

Example:

High electricity usage at 3 AM.

High temperature in winter season.

Key Points:

Context is crucial.

Normal in one context, abnormal in another.

Common in time-series data.

Used in IoT and energy monitoring.

Requires contextual features.

3. Collective Anomalies

A group of data points that together form an anomaly.

Example:

Sudden continuous drop in stock prices.

Repeated failed login attempts.

Key Points:

Individual points may appear normal.

Pattern as a whole is abnormal.

Harder to detect.

Common in network intrusion.

Requires sequence analysis.

Question 3: What are the key components of a Time Series? Explain each with one example.

Answer:

A Time Series is a sequence of observations recorded over time.

Key Components:
1. Trend

Long-term movement in data.

Example:
Increasing airline passengers over years.

2. Seasonality

Repeating patterns at fixed intervals.

Example:
Higher sales during festivals every year.

3. Cyclical Component

Fluctuations without fixed period.

Example:
Economic boom and recession cycles.

4. Residual (Noise)

Random fluctuations.

Example:
Unexpected weather disruptions.

Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

Answer:

A time series is stationary if its statistical properties remain constant over time.

Characteristics of Stationary Series:

Constant mean.

Constant variance.

Constant autocorrelation.

Testing Stationarity:

Augmented Dickey-Fuller (ADF) test.

Visual inspection.

Rolling statistics.

KPSS test.

Autocorrelation plots.

Transforming Non-Stationary Series:

Differencing.

Log transformation.

Seasonal differencing.

Moving average smoothing.

Box-Cox transformation.
